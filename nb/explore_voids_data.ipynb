```python
import numpy as np
import matplotlib.pyplot as plt
from pathlib import Path
%matplotlib inline
%config InlineBackend.figure_format = 'retina'

cosmo = 126

data_dir = Path('/pscratch/sd/e/epaillas/sunbird/data/clustering/abacus_cutsky/voidprior/voids/')
data_fn = data_dir / f'voids_multipoles_c{cosmo:03}_ph000.npy'
data = np.load(data_fn, allow_pickle=True).item()

s = data['s']
multipoles = data['multipoles']

fig, ax = plt.subplots(1, 2, figsize=(10, 4))

for i in range(len(multipoles)):
    ax[0].plot(s, multipoles[i, 0])
    ax[1].plot(s, multipoles[i, 1])


plt.tight_layout()
plt.show()
```
